```python
input_user = input("Masukkan bilangan: ")

try:
    bilangan = int(input_user)

    if bilangan > 0:
        print("Positif")
    elif bilangan < 0:
        print("Negatif")
    else:
        print("Nol")

except:
    print("Input tidak valid! Harus berupa angka.")
```

```python
input_bulan = input("Masukkan bulan (1-12): ")

try:
    bulan = int(input_bulan)

    if bulan == 2:
        print("Jumlah hari: 29")
    elif bulan in [4, 6, 9, 11]:
        print("Jumlah hari: 30")
    elif bulan in [1, 3, 5, 7, 8, 10, 12]:
        print("Jumlah hari: 31")
    else:
        print("Bulan tidak valid!")

except:
    print("Input harus berupa angka!")
```

```python
try:
    s1 = int(input("Masukkan sisi 1: "))
    s2 = int(input("Masukkan sisi 2: "))
    s3 = int(input("Masukkan sisi 3: "))

    if s1 == s2 == s3:
        print("3 sisi sama")
    elif s1 == s2 or s1 == s3 or s2 == s3:
        print("2 sisi sama")
    else:
        print("Tidak ada yang sama")

except:
    print("Input tidak valid! Harus berupa angka.")
```
