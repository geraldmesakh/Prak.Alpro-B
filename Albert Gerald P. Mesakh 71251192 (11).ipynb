```python
def tiga_terbaik(data):
    data_urut = sorted(data, reverse=True)
    
    return data_urut[:3]


angka = [10, 3, 25, 7, 18, 2]
hasil = tiga_terbaik(angka)

print("3 bilangan terbaik:", hasil)
```

```python
total = 0
count = 0

while True:
    user_input = input("Masukkan angka (atau ketik 'done'): ")
    
    if user_input == "done":
        break
    
    try:
        angka = float(user_input)
        total += angka
        count += 1
    except:
        print("Input harus berupa angka!")

if count > 0:
    rata_rata = total / count
    print("Rata-rata:", rata_rata)
else:
    print("Tidak ada data yang dimasukkan.")
```

```python
filename = input("Masukkan nama file: ")

try:
    handle = open(filename)
    kata_unik = set()

    for line in handle:
        line = line.strip().lower()
        words = line.split()

        for kata in words:
            kata_unik.add(kata)

    handle.close()

    hasil = sorted(kata_unik)

    print("Kata-kata unik dalam file:")
    for kata in hasil:
        print(kata)

except:
    print("File tidak ditemukan!")
```
