```python
data = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

print("key\tvalue\titem")

for key, value in data.items():
    print(key, "\t", value, "\t", key)
```

```python
lista = ['red', 'green', 'blue']
listb = ['#FF0000', '#008000', '#0000FF']

data = dict(zip(lista, listb))

hasil = {
    'green': data['green'],
    'blue': data['blue'],
    'red': data['red']
}

print(hasil)
```

```python
nama_file = input("Masukkan nama file : ")

file = open(nama_file)

data = {}

for baris in file:
    if baris.startswith("From "):
        kata = baris.split()
        email = kata[1]

        data[email] = data.get(email, 0) + 1

print(data)
```

```python
nama_file = input("Masukkan nama file : ")

file = open(nama_file)

domain_count = {}

for baris in file:
    if baris.startswith("From "):
        kata = baris.split()
        email = kata[1]

        domain = email.split("@")[1]

        domain_count[domain] = domain_count.get(domain, 0) + 1

print(domain_count)
```
