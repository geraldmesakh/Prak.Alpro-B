```python
def cek_angka(a, b, c):
    if a != b and b != c and a != c:
        if a + b == c or a + c == b or b + c == a:
            return True
        else:
            return False
    else:
        return False
print(cek_angka(1, 2, 3))
print(cek_angka(5, 5, 10))
print(cek_angka(10, 5, 3))
```

```python
celcius_ke_fahrenheit = lambda C: (9/5) * C + 32
celcius_ke_reamur = lambda C: 0.8 * C

C = float(input("Input C = "))
print("Output F =", celcius_ke_fahrenheit(C))

C = float(input("Input C = "))
print("Output R =", celcius_ke_reamur(C))

C = float(input("Input = "))
print("Output F =", celcius_ke_fahrenheit(C))
```
