```python
def is_prima(x):
    if x < 2:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return False
    return True

n = int(input("Masukkan n: "))

for i in range(n-1, 1, -1):
    if is_prima(i):
        print("Bilangan prima terdekat < n adalah:", i)
        break
```

```python
import math

n = int(input("Masukkan n: "))

for i in range(n, 0, -1):
    print(math.factorial(i), end=" ")
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```
