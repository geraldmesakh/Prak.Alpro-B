```python
jumlah = int(input("Masukkan jumlah mata kuliah: "))

total = 0

for i in range(jumlah):
    nilai = input("Masukkan nilai (A/B/C/D): ")

    if nilai == "A":
        bobot = 4
    elif nilai == "B":
        bobot = 3
    elif nilai == "C":
        bobot = 2
    elif nilai == "D":
        bobot = 1

    total = total + (bobot * 3)

ips = total / (jumlah * 3)

print("Hasil IPS yang didapatkan adalah=", ips)
```
