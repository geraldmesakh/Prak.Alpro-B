```python
data = ('Gerald Mesakh', '71251192', 'Sagan, DI Yogyakarta')

print("NIM    :", data[1])
print("NAMA   :", data[0])
print("ALAMAT :", data[2])

nim_tuple = tuple(data[1])

nama_depan = tuple(data[0].split()[0].lower())

nama_terbalik = tuple(reversed(data[0].split()))

print("\nNIM:", nim_tuple)
print("\nNAMA DEPAN:", nama_depan)
print("\nNAMA TERBALIK:", nama_terbalik)
```

```python
nama_file = input("Enter a file name: ")

file = open(nama_file)

jam_count = {}

for baris in file:
    if baris.startswith("From "):

        kata = baris.split()

        waktu = kata[5]

        jam = waktu.split(":")[0]

        # Menghitung jumlah kemunculan jam
        jam_count[jam] = jam_count.get(jam, 0) + 1

# Menampilkan hasil urut
for jam, jumlah in sorted(jam_count.items()):
    print(jam, jumlah)
```
